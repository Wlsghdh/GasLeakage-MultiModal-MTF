```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import normaltest, shapiro, anderson
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import warnings
warnings.filterwarnings('ignore')

# í•œê¸€ í°íŠ¸ ì„¤ì •
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['axes.unicode_minus'] = False

# ë°ì´í„° ë¡œë“œ
sensor = pd.read_csv('/data2/project/2025summer/jjh0709/Data/zkwgkjkjn9-2/Gas Sensors Measurements/Gas_Sensors_Measurements.csv')

print("="*80)
print("GAS SENSOR DATA - í¬ê´„ì  íƒìƒ‰ì  ë°ì´í„° ë¶„ì„ (EDA)")
print("="*80)

# ===== 1. ê¸°ë³¸ ë°ì´í„° ì •ë³´ =====
print("\n1. ê¸°ë³¸ ë°ì´í„° ì •ë³´")
print("-"*50)
print(f"ë°ì´í„° í˜•íƒœ: {sensor.shape}")
print(f"ë©”ëª¨ë¦¬ ì‚¬ìš©ëŸ‰: {sensor.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
print("\në°ì´í„° íƒ€ìž…:")
print(sensor.dtypes)
print("\nì²« 5í–‰:")
print(sensor.head())
print("\në§ˆì§€ë§‰ 5í–‰:")
print(sensor.tail())

# ===== 2. ê¸°ìˆ  í†µê³„ëŸ‰ =====
print("\n\n2. ê¸°ìˆ  í†µê³„ëŸ‰")
print("-"*50)
numeric_cols = sensor.select_dtypes(include=[np.number]).columns.tolist()
sensor_cols = [col for col in numeric_cols if col.startswith('MQ')]

print("ì „ì²´ ìˆ˜ì¹˜í˜• ë³€ìˆ˜ ê¸°ìˆ í†µê³„:")
print(sensor[numeric_cols].describe())

print(f"\nì„¼ì„œë³„ ì„¸ë¶€ í†µê³„ (MQ ì„¼ì„œ: {len(sensor_cols)}ê°œ):")
for col in sensor_cols:
    data = sensor[col]
    print(f"\n{col}:")
    print(f"  í‰ê· : {data.mean():.2f}")
    print(f"  ì¤‘ì•™ê°’: {data.median():.2f}")
    print(f"  í‘œì¤€íŽ¸ì°¨: {data.std():.2f}")
    print(f"  ë³€ë™ê³„ìˆ˜: {data.std()/data.mean():.3f}")
    print(f"  ìµœì†Ÿê°’: {data.min()}")
    print(f"  ìµœëŒ“ê°’: {data.max()}")
    print(f"  ë²”ìœ„: {data.max() - data.min()}")
    print(f"  ì‚¬ë¶„ìœ„ìˆ˜ ë²”ìœ„(IQR): {data.quantile(0.75) - data.quantile(0.25):.2f}")
    print(f"  ì™œë„: {stats.skew(data):.3f}")
    print(f"  ì²¨ë„: {stats.kurtosis(data):.3f}")

# ===== 3. íƒ€ê²Ÿ ë³€ìˆ˜ ë¶„ì„ =====
print("\n\n3. íƒ€ê²Ÿ ë³€ìˆ˜ (Gas) ë¶„ì„")
print("-"*50)
gas_counts = sensor['Gas'].value_counts()
gas_props = sensor['Gas'].value_counts(normalize=True)

print("ê°€ìŠ¤ íƒ€ìž…ë³„ ë¶„í¬:")
for gas_type, count in gas_counts.items():
    print(f"  {gas_type}: {count}ê°œ ({gas_props[gas_type]*100:.1f}%)")

print(f"\ní´ëž˜ìŠ¤ ë¶ˆê· í˜• ë¹„ìœ¨: {gas_counts.max()/gas_counts.min():.2f}:1")

# ===== 4. ì‹œê°í™” ì‹œìž‘ =====
print("\n\n4. ì‹œê°í™” ë¶„ì„ ì‹œìž‘")
print("-"*50)

# 4-1. íƒ€ê²Ÿ ë³€ìˆ˜ ë¶„í¬ ì‹œê°í™”
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# ë§‰ëŒ€ ê·¸ëž˜í”„
gas_counts.plot(kind='bar', ax=axes[0], color='skyblue', edgecolor='black')
axes[0].set_title('Gas Type Distribution', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Gas Type')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=45)

# íŒŒì´ ì°¨íŠ¸
axes[1].pie(gas_counts.values, labels=gas_counts.index, autopct='%1.1f%%', 
           colors=['lightcoral', 'lightblue', 'lightgreen', 'gold'])
axes[1].set_title('Gas Type Proportion', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

# 4-2. ì„¼ì„œë³„ ë¶„í¬ ì‹œê°í™” (ížˆìŠ¤í† ê·¸ëž¨)
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.ravel()

for i, col in enumerate(sensor_cols):
    axes[i].hist(sensor[col], bins=50, alpha=0.7, color='lightblue', edgecolor='black')
    axes[i].set_title(f'{col} Distribution', fontweight='bold')
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 4-3. ë°•ìŠ¤í”Œë¡¯ - ì„¼ì„œë³„ ë¶„í¬
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.ravel()

for i, col in enumerate(sensor_cols):
    axes[i].boxplot(sensor[col])
    axes[i].set_title(f'{col} Box Plot', fontweight='bold')
    axes[i].set_ylabel('Value')
    axes[i].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 4-4. ê°€ìŠ¤ íƒ€ìž…ë³„ ì„¼ì„œ ê°’ ë¶„í¬ (ë°•ìŠ¤í”Œë¡¯)
fig, axes = plt.subplots(2, 4, figsize=(20, 12))
axes = axes.ravel()

for i, col in enumerate(sensor_cols):
    sensor.boxplot(column=col, by='Gas', ax=axes[i])
    axes[i].set_title(f'{col} by Gas Type')
    axes[i].set_xlabel('Gas Type')
    axes[i].set_ylabel('Value')

plt.suptitle('Sensor Values Distribution by Gas Type', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# 4-5. ë°”ì´ì˜¬ë¦° í”Œë¡¯
fig, axes = plt.subplots(2, 4, figsize=(20, 12))
axes = axes.ravel()

for i, col in enumerate(sensor_cols):
    sns.violinplot(data=sensor, x='Gas', y=col, ax=axes[i])
    axes[i].set_title(f'{col} Distribution by Gas Type')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# ===== 5. ìƒê´€ê´€ê³„ ë¶„ì„ =====
print("\n\n5. ìƒê´€ê´€ê³„ ë¶„ì„")
print("-"*50)

correlation_matrix = sensor[sensor_cols].corr()
print("ì„¼ì„œê°„ ìƒê´€ê´€ê³„ ë§¤íŠ¸ë¦­ìŠ¤:")
print(correlation_matrix.round(3))

# ë†’ì€ ìƒê´€ê´€ê³„ ìŒ ì°¾ê¸°
high_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        corr_val = correlation_matrix.iloc[i, j]
        if abs(corr_val) > 0.7:  # 0.7 ì´ìƒì˜ ìƒê´€ê´€ê³„
            high_corr_pairs.append((correlation_matrix.columns[i], 
                                   correlation_matrix.columns[j], 
                                   corr_val))

print(f"\në†’ì€ ìƒê´€ê´€ê³„ (|r| > 0.7) ìŒë“¤:")
for pair in high_corr_pairs:
    print(f"  {pair[0]} - {pair[1]}: {pair[2]:.3f}")

# ìƒê´€ê´€ê³„ ížˆíŠ¸ë§µ
plt.figure(figsize=(12, 10))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm', 
           center=0, square=True, fmt='.2f')
plt.title('Sensor Correlation Heatmap', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# ===== 6. ë‹¤ì¤‘ê³µì„ ì„± ë¶„ì„ (VIF) =====
print("\n\n6. ë‹¤ì¤‘ê³µì„ ì„± ë¶„ì„ (VIF)")
print("-"*50)

from statsmodels.stats.outliers_influence import variance_inflation_factor

X = sensor[sensor_cols].values
vif_data = pd.DataFrame()
vif_data["Feature"] = sensor_cols
vif_data["VIF"] = [variance_inflation_factor(X, i) for i in range(len(sensor_cols))]
vif_data = vif_data.sort_values('VIF', ascending=False)

print("ë¶„ì‚°íŒ½ì°½ì¸ìˆ˜ (VIF) ê²°ê³¼:")
print(vif_data)
print("\ní•´ì„:")
print("- VIF < 5: ë‹¤ì¤‘ê³µì„ ì„± ë¬¸ì œ ì—†ìŒ")
print("- 5 â‰¤ VIF < 10: ì¤‘ê°„ ì •ë„ì˜ ë‹¤ì¤‘ê³µì„ ì„±")
print("- VIF â‰¥ 10: ì‹¬ê°í•œ ë‹¤ì¤‘ê³µì„ ì„± ë¬¸ì œ")

high_vif = vif_data[vif_data['VIF'] > 10]
if not high_vif.empty:
    print(f"\në†’ì€ VIF (>10)ë¥¼ ê°€ì§„ ë³€ìˆ˜ë“¤:")
    for _, row in high_vif.iterrows():
        print(f"  {row['Feature']}: {row['VIF']:.2f}")

# ===== 7. ì •ê·œì„± ê²€ì • =====
print("\n\n7. ì •ê·œì„± ê²€ì •")
print("-"*50)

normality_results = []
for col in sensor_cols:
    data = sensor[col]
    
    # Shapiro-Wilk ê²€ì • (ìƒ˜í”Œ í¬ê¸°ê°€ í´ ë•ŒëŠ” ë¶€ì í•©í•˜ë¯€ë¡œ ìƒ˜í”Œë§)
    if len(data) > 5000:
        sample_data = data.sample(n=5000, random_state=42)
    else:
        sample_data = data
    
    shapiro_stat, shapiro_p = shapiro(sample_data)
    
    # Anderson-Darling ê²€ì •
    anderson_result = anderson(data, dist='norm')
    
    # Kolmogorov-Smirnov ê²€ì •
    ks_stat, ks_p = stats.kstest(data, 'norm', args=(data.mean(), data.std()))
    
    normality_results.append({
        'Sensor': col,
        'Shapiro_stat': shapiro_stat,
        'Shapiro_p': shapiro_p,
        'Anderson_stat': anderson_result.statistic,
        'KS_stat': ks_stat,
        'KS_p': ks_p,
        'Normal': 'Yes' if shapiro_p > 0.05 and ks_p > 0.05 else 'No'
    })

normality_df = pd.DataFrame(normality_results)
print(normality_df)

# Q-Q í”Œë¡¯
fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.ravel()

for i, col in enumerate(sensor_cols):
    stats.probplot(sensor[col], dist="norm", plot=axes[i])
    axes[i].set_title(f'{col} Q-Q Plot')

plt.tight_layout()
plt.show()

# ===== 8. ì´ìƒì¹˜ ë¶„ì„ =====
print("\n\n8. ì´ìƒì¹˜ ë¶„ì„")
print("-"*50)

outlier_summary = []
for col in sensor_cols:
    data = sensor[col]
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    outlier_count = len(outliers)
    outlier_percentage = (outlier_count / len(data)) * 100
    
    # Z-score ë°©ë²•
    z_scores = np.abs(stats.zscore(data))
    z_outliers = data[z_scores > 3]
    z_outlier_count = len(z_outliers)
    
    outlier_summary.append({
        'Sensor': col,
        'IQR_Outliers': outlier_count,
        'IQR_Percentage': outlier_percentage,
        'Z_Outliers': z_outlier_count,
        'Lower_Bound': lower_bound,
        'Upper_Bound': upper_bound
    })

outlier_df = pd.DataFrame(outlier_summary)
print("ì´ìƒì¹˜ ë¶„ì„ ê²°ê³¼:")
print(outlier_df)

# ===== 9. ê°€ìŠ¤ íƒ€ìž…ë³„ ì„¼ì„œ í†µê³„ =====
print("\n\n9. ê°€ìŠ¤ íƒ€ìž…ë³„ ì„¼ì„œ í†µê³„")
print("-"*50)

for gas_type in sensor['Gas'].unique():
    print(f"\n{gas_type} ê°€ìŠ¤:")
    gas_data = sensor[sensor['Gas'] == gas_type][sensor_cols]
    print(gas_data.describe())

# ê°€ìŠ¤ íƒ€ìž…ë³„ ì„¼ì„œ ê°’ í‰ê·  ë¹„êµ
gas_means = sensor.groupby('Gas')[sensor_cols].mean()
print("\nê°€ìŠ¤ íƒ€ìž…ë³„ ì„¼ì„œ í‰ê· ê°’:")
print(gas_means)

# ê°€ìŠ¤ íƒ€ìž…ë³„ í‰ê· ê°’ ížˆíŠ¸ë§µ
plt.figure(figsize=(12, 6))
sns.heatmap(gas_means.T, annot=True, cmap='viridis', fmt='.0f')
plt.title('Average Sensor Values by Gas Type', fontsize=16, fontweight='bold')
plt.xlabel('Gas Type')
plt.ylabel('Sensors')
plt.tight_layout()
plt.show()

# ===== 10. ì„¼ì„œê°„ ì‚°ì ë„ ë§¤íŠ¸ë¦­ìŠ¤ =====
print("\n\n10. ì„¼ì„œê°„ ê´€ê³„ ì‹œê°í™”")
print("-"*50)

# ì£¼ìš” ì„¼ì„œë“¤ì˜ ì‚°ì ë„ ë§¤íŠ¸ë¦­ìŠ¤ (ì²˜ìŒ 4ê°œ ì„¼ì„œ)
main_sensors = sensor_cols[:4]
sns.pairplot(sensor[main_sensors + ['Gas']], hue='Gas', diag_kind='hist')
plt.suptitle('Sensor Pairplot (First 4 Sensors)', y=1.02, fontsize=16, fontweight='bold')
plt.show()

# ===== 11. ìŠ¤ì¼€ì¼ë§ í•„ìš”ì„± ë¶„ì„ =====
print("\n\n11. ìŠ¤ì¼€ì¼ë§ í•„ìš”ì„± ë¶„ì„")
print("-"*50)

sensor_stats = sensor[sensor_cols].agg(['mean', 'std', 'min', 'max'])
sensor_stats.loc['range'] = sensor_stats.loc['max'] - sensor_stats.loc['min']
sensor_stats.loc['cv'] = sensor_stats.loc['std'] / sensor_stats.loc['mean']

print("ì„¼ì„œë³„ í†µê³„ (ìŠ¤ì¼€ì¼ë§ íŒë‹¨ìš©):")
print(sensor_stats.round(2))

print("\nìŠ¤ì¼€ì¼ë§ í•„ìš”ì„± íŒë‹¨:")
print("1. í‘œì¤€íŽ¸ì°¨ ë¹„êµ:")
std_ratio = sensor_stats.loc['std'].max() / sensor_stats.loc['std'].min()
print(f"   ìµœëŒ€/ìµœì†Œ í‘œì¤€íŽ¸ì°¨ ë¹„ìœ¨: {std_ratio:.2f}")

print("2. ë²”ìœ„ ë¹„êµ:")
range_ratio = sensor_stats.loc['range'].max() / sensor_stats.loc['range'].min()
print(f"   ìµœëŒ€/ìµœì†Œ ë²”ìœ„ ë¹„ìœ¨: {range_ratio:.2f}")

print("3. ë³€ë™ê³„ìˆ˜ ë¹„êµ:")
cv_diff = sensor_stats.loc['cv'].max() - sensor_stats.loc['cv'].min()
print(f"   ë³€ë™ê³„ìˆ˜ ì°¨ì´: {cv_diff:.3f}")

if std_ratio > 10 or range_ratio > 10:
    print("\nê¶Œìž¥ì‚¬í•­: ìŠ¤ì¼€ì¼ë§ì´ í•„ìš”í•©ë‹ˆë‹¤!")
    print("- StandardScaler ë˜ëŠ” MinMaxScaler ì‚¬ìš© ê¶Œìž¥")
else:
    print("\nê¶Œìž¥ì‚¬í•­: ìŠ¤ì¼€ì¼ë§ì´ ì„ íƒì‚¬í•­ìž…ë‹ˆë‹¤.")

# ìŠ¤ì¼€ì¼ë§ ì „í›„ ë¹„êµ ì‹œê°í™”
scaler_standard = StandardScaler()
scaler_minmax = MinMaxScaler()

sensor_scaled_std = pd.DataFrame(
    scaler_standard.fit_transform(sensor[sensor_cols]),
    columns=sensor_cols
)

sensor_scaled_mm = pd.DataFrame(
    scaler_minmax.fit_transform(sensor[sensor_cols]),
    columns=sensor_cols
)

fig, axes = plt.subplots(3, 1, figsize=(15, 12))

# ì›ë³¸ ë°ì´í„°
sensor[sensor_cols].boxplot(ax=axes[0])
axes[0].set_title('Original Data', fontweight='bold')
axes[0].set_ylabel('Values')

# Standard Scaling
sensor_scaled_std.boxplot(ax=axes[1])
axes[1].set_title('After Standard Scaling', fontweight='bold')
axes[1].set_ylabel('Standardized Values')

# MinMax Scaling
sensor_scaled_mm.boxplot(ax=axes[2])
axes[2].set_title('After MinMax Scaling', fontweight='bold')
axes[2].set_ylabel('Normalized Values')

plt.tight_layout()
plt.show()

# ===== 12. ì£¼ì„±ë¶„ ë¶„ì„ (PCA) ë¯¸ë¦¬ë³´ê¸° =====
print("\n\n12. ì£¼ì„±ë¶„ ë¶„ì„ (PCA) ë¯¸ë¦¬ë³´ê¸°")
print("-"*50)

from sklearn.decomposition import PCA

# í‘œì¤€í™”ëœ ë°ì´í„°ë¡œ PCA
pca = PCA()
pca_result = pca.fit_transform(sensor_scaled_std)

# ì„¤ëª…ëœ ë¶„ì‚° ë¹„ìœ¨
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

print("ì£¼ì„±ë¶„ë³„ ì„¤ëª…ëœ ë¶„ì‚° ë¹„ìœ¨:")
for i, ratio in enumerate(explained_variance_ratio):
    print(f"  PC{i+1}: {ratio:.3f} ({ratio*100:.1f}%)")

print(f"\nëˆ„ì  ì„¤ëª…ëœ ë¶„ì‚°:")
for i, cum_ratio in enumerate(cumulative_variance_ratio):
    print(f"  PC1~PC{i+1}: {cum_ratio:.3f} ({cum_ratio*100:.1f}%)")

# ìŠ¤í¬ë¦¬ í”Œë¡¯
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

axes[0].plot(range(1, len(explained_variance_ratio)+1), explained_variance_ratio, 'bo-')
axes[0].set_title('Scree Plot', fontweight='bold')
axes[0].set_xlabel('Principal Component')
axes[0].set_ylabel('Explained Variance Ratio')
axes[0].grid(True, alpha=0.3)

axes[1].plot(range(1, len(cumulative_variance_ratio)+1), cumulative_variance_ratio, 'ro-')
axes[1].axhline(y=0.95, color='k', linestyle='--', alpha=0.5)
axes[1].set_title('Cumulative Explained Variance', fontweight='bold')
axes[1].set_xlabel('Principal Component')
axes[1].set_ylabel('Cumulative Explained Variance Ratio')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ===== 13. ìµœì¢… ìš”ì•½ =====
print("\n\n" + "="*80)
print("EDA ìµœì¢… ìš”ì•½")
print("="*80)

print(f"\në°ì´í„° ê°œìš”:")
print(f"- ì´ ìƒ˜í”Œ ìˆ˜: {len(sensor):,}ê°œ")
print(f"- ì„¼ì„œ ê°œìˆ˜: {len(sensor_cols)}ê°œ")
print(f"- ê°€ìŠ¤ íƒ€ìž…: {len(sensor['Gas'].unique())}ê°œ")

print(f"\ní´ëž˜ìŠ¤ ë¶„í¬:")
for gas_type, count in gas_counts.items():
    print(f"- {gas_type}: {count:,}ê°œ ({count/len(sensor)*100:.1f}%)")

print(f"\në°ì´í„° í’ˆì§ˆ:")
print(f"- ê²°ì¸¡ê°’: ì—†ìŒ")
print(f"- ì¤‘ë³µê°’: í™•ì¸ ë¶ˆí•„ìš” (ì‚¬ì „ ì •ë³´)")

print(f"\nì£¼ìš” ë°œê²¬ì‚¬í•­:")
if high_corr_pairs:
    print(f"- ë†’ì€ ìƒê´€ê´€ê³„ ìŒ: {len(high_corr_pairs)}ê°œ")
if not high_vif.empty:
    print(f"- ë‹¤ì¤‘ê³µì„ ì„± ë¬¸ì œ ë³€ìˆ˜: {len(high_vif)}ê°œ")

print(f"\nì •ê·œì„±:")
normal_sensors = normality_df[normality_df['Normal'] == 'Yes']['Sensor'].tolist()
print(f"- ì •ê·œë¶„í¬ë¥¼ ë”°ë¥´ëŠ” ì„¼ì„œ: {len(normal_sensors)}ê°œ / {len(sensor_cols)}ê°œ")

print(f"\nì´ìƒì¹˜:")
total_outliers = outlier_df['IQR_Outliers'].sum()
print(f"- ì´ ì´ìƒì¹˜ ê°œìˆ˜ (IQR ë°©ë²•): {total_outliers:,}ê°œ")

print(f"\nìŠ¤ì¼€ì¼ë§ ê¶Œìž¥ì‚¬í•­:")
if std_ratio > 10 or range_ratio > 10:
    print("- ìŠ¤ì¼€ì¼ë§ í•„ìš”í•¨ (í‘œì¤€íŽ¸ì°¨/ë²”ìœ„ ì°¨ì´ê°€ í¼)")
else:
    print("- ìŠ¤ì¼€ì¼ë§ ì„ íƒì‚¬í•­")

print(f"\nì°¨ì› ì¶•ì†Œ:")
pcs_for_95 = np.argmax(cumulative_variance_ratio >= 0.95) + 1
print(f"- 95% ë¶„ì‚° ì„¤ëª…ì„ ìœ„í•œ ì£¼ì„±ë¶„ ê°œìˆ˜: {pcs_for_95}ê°œ / {len(sensor_cols)}ê°œ")

print("\n" + "="*80)
print("EDA ì™„ë£Œ!")
print("="*80)
```
