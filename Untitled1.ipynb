```python
import pandas as pd
import numpy as np


sensor = pd.read_csv("/data2/project/2025summer/yjh0913/gasdata/Gas Sensors Measurements/Gas_Sensors_Measurements.csv")
sensor.rename(columns={sensor.columns[0]: "SerialNumber"}, inplace=True)
sensor['Gas'] = sensor['Gas'].str.strip()


sensor_columns = ['MQ2', 'MQ3', 'MQ5', 'MQ6', 'MQ7', 'MQ8', 'MQ135']


target_class = "Smoke"
filtered = sensor[sensor['Gas'] == target_class].copy()


intervals = [(10, 44), (45, 80), (81, 190), (191, 330), (331, 440), (441, 515), (516, 600), (601, 700), 
             (701, 825), (826, 948), (949, 995), (996, 1075), (1076, 1170), (1171, 1250), (1251, 1340),
             (1341, 1397), (1398, 1460), (1461, 1530), (1531, 1600)]  #    

normalized_segments = []

#   
for start, end in intervals:
    segment = filtered[(filtered['SerialNumber'] >= start) & (filtered['SerialNumber'] <= end)].copy()
    
    if not segment.empty:
        # Min-Max 
        normalized = (segment[sensor_columns] - segment[sensor_columns].min()) / \
                     (segment[sensor_columns].max() - segment[sensor_columns].min())
        
        #   
        normalized['SerialNumber'] = segment['SerialNumber']
        normalized['Gas'] = segment['Gas']
        normalized['Interval'] = f"{start}-{end}"  #   
        
        normalized_segments.append(normalized)

#  
normalized_data = pd.concat(normalized_segments, ignore_index=True)

#   
print(normalized_data.head())

#   :
normalized_data.to_csv("normalized_Smoke.csv", index=False)
```

```python
print("\n8.  ")
print("-"*50)

outlier_summary = []
for col in sensor_cols:
    data = sensor[col]
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    outlier_count = len(outliers)
    outlier_percentage = (outlier_count / len(data)) * 100
    
    # Z-score 
    z_scores = np.abs(stats.zscore(data))
    z_outliers = data[z_scores > 3]
    z_outlier_count = len(z_outliers)
    
    outlier_summary.append({
        'Sensor': col,
        'IQR_Outliers': outlier_count,
        'IQR_Percentage': outlier_percentage,
        'Z_Outliers': z_outlier_count,
        'Lower_Bound': lower_bound,
        'Upper_Bound': upper_bound
    })

outlier_df = pd.DataFrame(outlier_summary)
print("  :")
print(outlier_df)
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split

sensor_cols = ['MQ2','MQ3','MQ5','MQ6','MQ7','MQ8','MQ135']
label_col = 'Gas'

le = LabelEncoder()
y = le.fit_transform(sensor[label_col])

X_train, X_test, y_train, y_test = train_test_split(sensor[sensor_cols], y, test_size=0.2, stratify=y, random_state=42)
print(X_train.shape, y_train.shape,X_test.shape)
model = ExtraTreesClassifier(
    n_estimators=100,     
    max_features=2,         
    random_state=42,
    n_jobs=-1               
)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1.  
df = pd.read_csv('/data2/project/2025summer/jjh0709/Data/zkwgkjkjn9-2/Gas Sensors Measurements/Gas_Sensors_Measurements.csv')

# 2. 'Mixture'   MQ8   
mixture_mq8 = df[df['Gas'] == 'Mixture'][['Serial Number', 'MQ8']].reset_index(drop=True)

# 3.      ( 0 )
gradient = np.gradient(mixture_mq8['MQ8'].values)
zero_slope_indices = np.where(gradient == 0)[0]

# 4.     0    
start_indices = zero_slope_indices + 1
start_indices = start_indices[start_indices < len(mixture_mq8)]

# 5.   :   100,  20 
new_values_with_padding = []
i = 0
while i < len(mixture_mq8):
    next_start = start_indices[start_indices > i]
    if len(next_start) == 0:
        segment = mixture_mq8['MQ8'].values[i:]
        new_values_with_padding.extend([segment[0]] * 100)  #  100
        new_values_with_padding.extend(segment.tolist())    # 
        break
    end = next_start[0]
    segment = mixture_mq8['MQ8'].values[i:end]
    new_values_with_padding.extend([segment[0]] * 100)      #  100
    new_values_with_padding.extend(segment.tolist())        # 
    new_values_with_padding.extend([segment[-1]] * 20)      #  20
    i = end

# 6.  CSV 
result_df = pd.DataFrame({'MQ8_Mixture_Timeseries': new_values_with_padding})
result_df.to_csv('MQ8_Mixture_Timeseries_with_padding.csv', index=False)

# 7. 
plt.figure(figsize=(15, 5))
plt.plot(new_values_with_padding, label='MQ8 (Mixture)', linewidth=1)
plt.title('MQ8 Sensor Time Series with Padding (Mixture)')
plt.xlabel('Time (samples)')
plt.ylabel('MQ8 Sensor Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
